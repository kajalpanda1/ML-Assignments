```
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
import sklearn.metrics as sm
```

```
data=pd.read_csv('https://raw.githubusercontent.com/kajalpanda1/DatasetRepository/master/Balance.csv',header=None)
```

```
data.head()
```

```
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data[0]=le.fit_transform(data[0])
data.head()
```

```
from sklearn.preprocessing import Normalizer
norm=Normalizer()
data_norm=norm.fit_transform(data)
```

```
from sklearn.decomposition import PCA
p=PCA(n_components=2)
X=p.fit_transform(data_norm)
```

```
from scipy.cluster.hierarchy import dendrogram, linkage
link_matrix=linkage(X,method='single')
```

```
plt.figure(figsize=(10,10))
dendrogram(link_matrix)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Cluster Size')
plt.ylabel('Distance')
plt.show()
```

```
cluster=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='single')
cluster.fit(X)
labels=cluster.labels_

plt.scatter(X[labels==0, 0], X[labels==0, 1], s=50, marker='o', color='cyan')
plt.scatter(X[labels==1, 0], X[labels==1, 1], s=50, marker='o', color='green')
plt.scatter(X[labels==2, 0], X[labels==2, 1], s=50, marker='o', color='yellow')
```

```
cluster=AgglomerativeClustering(n_clusters=2)
plt.figure(figsize=(8, 5))
plt.scatter(X[:,0], X[:,1], c=cluster.fit_predict(X), cmap='rainbow')

cluster=AgglomerativeClustering(n_clusters=3)
plt.figure(figsize=(8, 5))
plt.scatter(X[:,0], X[:,1], c=cluster.fit_predict(X), cmap='rainbow')

cluster=AgglomerativeClustering(n_clusters=4)
plt.figure(figsize=(8, 5))
plt.scatter(X[:,0], X[:,1], c=cluster.fit_predict(X), cmap='rainbow')

cluster=AgglomerativeClustering(n_clusters=5)
plt.figure(figsize=(8, 5))
plt.scatter(X[:,0], X[:,1], c=cluster.fit_predict(X), cmap='rainbow')
```

```
from sklearn.metrics import silhouette_score 
for n in [2,3,4,5,6]:
    cluster=AgglomerativeClustering(n_clusters=n,linkage='single') 
    silhouette_avg=silhouette_score(X, cluster.fit_predict(X))
    print("For n_clusters =", n,"The average silhouette_score is :", round(silhouette_avg,2))
```

```
from sklearn.metrics import silhouette_score 
k = [2, 3, 4, 5, 6]
sil_scr = [] 

for x in k:
    ac = AgglomerativeClustering(n_clusters = x,linkage='complete') 
    sil_scr.append(silhouette_score(X, ac.fit_predict(X)))
plt.bar(k, sil_scr) 
plt.xlabel('Number of clusters', fontsize = 20) 
plt.ylabel('Silhouette Scores', fontsize = 20) 
plt.show()
```
